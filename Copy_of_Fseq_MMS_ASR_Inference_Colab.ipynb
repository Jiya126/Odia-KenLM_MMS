```python
#using kenLM and MMS on english
decoding_cmds = """
decoding.type=kenlm
decoding.beam=500
decoding.beamsizetoken=50
decoding.lmweight=2.69
decoding.wordscore=2.8
decoding.lmpath=/content/5gram_test_eng.bin
decoding.lexicon=/content/lexicon_eng.txt
""".replace("\n", " ")
!python examples/mms/asr/infer/mms_infer.py --model "/content/fairseq/models_new/mms1b_fl102.pt" --lang "eng" --audio '/content/male16.wav' \
    --extra-infer-args '{decoding_cmds}'
```
